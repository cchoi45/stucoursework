```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set random seed for reproducibility
np.random.seed(42)

# make non-normally distributed population
population = np.random.exponential(scale=2.0, size=100005)

# 1000 samples of size 30
sample_means = [np.mean(np.random.choice(population, size=30, replace=True)) for _ in range(1000)]

# Fit a normal distribution to the sample means
mu, sigma = np.mean(sample_means), np.std(sample_means)
x = np.linspace(min(sample_means), max(sample_means), 100)
normal_curve = stats.norm.pdf(x, mu, sigma)

# Plot the sampling distribution of the sample means
plt.figure(figsize=(9, 5))
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='navy', edgecolor='black', label='Sample Means')
plt.plot(x, normal_curve, 'k-', linewidth=2, label='Normal Distribution Fit')

# Labels and title
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('QUEST 3: Sampling Means Distribution with Sample Size = 30')
plt.legend()
plt.grid(axis='y', alpha=0.5)

# Show the plot
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set random seed for reproducibility
np.random.seed(42)

# Make non-normally distributed population
population = np.random.exponential(scale=2.0, size=100005)

# 1000 samples of size 30
sample_means = [np.mean(np.random.choice(population, size=30, replace=True)) for _ in range(1000)]

# Fit a normal distribution to the sample means
mu, sigma = np.mean(sample_means), np.std(sample_means)
x = np.linspace(min(sample_means), max(sample_means), 100)
normal_curve = stats.norm.pdf(x, mu, sigma)

# Create subplots to compare original population and sampling distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot the original population distribution
axes[0].hist(population, bins=50, density=True, alpha=0.7, color='green', edgecolor='black')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Original Population Distribution (Exponential)')
axes[0].grid(axis='y', alpha=0.5)

# Plot the sampling distribution of the sample means
axes[1].hist(sample_means, bins=30, density=True, alpha=0.7, color='navy', edgecolor='black', label='Sample Means')
axes[1].plot(x, normal_curve, 'k-', linewidth=2, label='Normal Distribution Fit')
axes[1].set_xlabel('Sample Mean')
axes[1].set_ylabel('Density')
axes[1].set_title('QUEST 3: Sampling Means Distribution with Sample Size = 30')
axes[1].legend()
axes[1].grid(axis='y', alpha=0.5)

# Show the plot
plt.tight_layout()
plt.show()
```
